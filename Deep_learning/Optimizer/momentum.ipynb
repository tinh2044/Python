```python
import numpy as np
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_samples=100,n_features=2, random_state=1, n_targets=1)
y = y.reshape(y.shape[0], 1)
```

```python
plt.scatter(X[:,1], y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
def mse(y, y_pred):
    return np.mean((y-y_pred)**2)
def mse_derivative(y, y_pred):
    return (1/len(y))*(y-y_pred)
```

```python
n_feat = X.shape[1]
n_samples = X.shape[0]
weights = np.zeros((n_feat, 1))
biases = 0
beta =  0.9
lr = 0.1
v_dW, v_dB = 0, 0
for i in range(10000):
    y_pred = X_train@weights +biases
    v_dW = beta*v_dW + lr*X_train.T@mse_derivative(y_train, y_pred)
    v_dB = beta*v_dB + lr*np.sum(mse_derivative(y_train, y_pred))
    
    # print(v_dW.shape)
    weights += v_dW
    biases += v_dB
    print(mse(y_train, y_pred))
```

```python
predict = X_test@weights +biases
```
