```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numba.cuda import grid
```

```python
#data set in sea seaborn
sns.get_dataset_names()
```

```python
data = sns.load_dataset('tips')
```

```python
data
```

```python
sns.histplot(data=data['total_bill'], cbar=True)
```

```python
sns.kdeplot(data=data['total_bill'])
```

```python
sns.displot(data=data, x='total_bill', kde=True, hue='sex', col='time')
```

```python

sns.barplot(data=data, x='day', y='tip', hue='sex')
```

```python
sns.countplot(data=data, x='smoker', hue='sex')
```

```python
sns.boxplot(data=data, x='day', y='total_bill', hue='sex')
```

```python
plt.figure(figsize=[10,5])
grid_plot = sns.FacetGrid(data= data , col='time', hue='day', height=3, aspect=1.5)
grid_plot.map(sns.scatterplot, 'tip', 'total_bill', s=50, edgecolor='b', alpha=0.6)
grid_plot.add_legend()
```
